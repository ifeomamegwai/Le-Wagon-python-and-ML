```python
# your code here
df_sales_restaurant_1['restaurant_id'] = 1

# Add 'restaurant_id' column to df_sales_restaurant_2 with id = 2
df_sales_restaurant_2['restaurant_id'] = 2

# Concatenate df_sales_restaurant_1 and df_sales_restaurant_2 along the rows (axis=0)
df_sales = pd.concat([df_sales_restaurant_1, df_sales_restaurant_2], axis=0)

# Display the result
print(df_sales)
```

## Working with an unidentified format


Someone from the team found a file named [`secret.pkl`](https://drive.google.com/file/d/1_oia73SoCZtZFsGbDrsffkSfdaj2Gt5O/view?usp=sharing).

Can you help them open it and find what's inside? Stackoverflow might have some answers!

```python
# your code here
import pickle

# Open the .pkl file in binary read mode
with open('secret.pkl', 'rb') as file:
    data = pickle.load(file)

# Inspect the contents
print(type(data))  # Check the type of the data
print(data)        # Print the actual data
```

Note: if you want to open your file into a dataframe, you can usually just use pandas!
