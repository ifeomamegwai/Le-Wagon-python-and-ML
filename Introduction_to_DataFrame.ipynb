### Masking

When we want to filter the rows of a DataFrame, we use a mask also called boolean masking.

```python
df.head()
```

```python
# first we create a condition  according the filtering we wan't to make.
# here, we would like to only keep the store_id 23.
df["store_id"] == 13 #--> result is a Series of boolean values (True/False)
```

```python
df["store_id"] > 11
```

```python
# this is called a boolean mask.
boolean_mask = (df["store_id"] == 13)
boolean_mask
```

Then we apply this mask to keep only the lines equal to True

```python
# we use the brackets to select the lines according to the mask.
df[boolean_mask]
```

```python
df[df["store_id"] == 13] #--> the same result, just without saving the boolean mask into a variable first
```

Multiple conditions can be combined in a mask

```python
(df["store_id"] == 23) | (df["item_id"] == 56)
```

```python
mask = (df["store_id"] == 23) | (df["item_id"] == 56)
df[mask]
```

```python
#& --> AND --> True AND True == True
#| --> OR --> True OR True == True ; True OR False == True
```

& means "and", | means "or"

### Groupby

You can use GROUP BY just like you did in SQL.
Use Groupby to resume info per dimension. In the following example, we calculate the total quantity per store.

1.   You mention first the name of your dataframe
2.   Then you add .groupby and you specify the name of the column you want to group on.
3.   Next, you put in brackets the columns you want to keep
4.   Lastly, you add the aggregation function you want to use.

```python
"""SELECT
  product_id,
  SUM (cost),
  AVG(turnover)
FROM table
group by product_id"""
```

```python
df.groupby("store_id")["quantity"].sum()
```

```python
df.groupby("store_id")["quantity"].mean()
```

Get the average for quantity and price at the same time.

Note, we don't need to use `.agg()` here because it's the same function we want apply.

```python
df.groupby("store_id")[["quantity", "price"]].mean()
```

You may also want to sort the output, you will need to use the `sort_values` function



```python
"""
SELECT
  store_id,
  AVERAGE(quantity) AS quantity,
  AVERAGE(price) AS price
FROM table
group by product_id
order by quantity DESC
"""
```

```python
df.groupby("store_id")[["quantity", "price"]].mean().sort_values(by=["quantity"], ascending=[False])
```

## Joins

### Concatenate

```python
df1 = pd.DataFrame({"id": [100, 101, 102],
                    "first_name": ["paul", "victor", "lucie"],
                    "last_name": ["oiseau", "blanc", "misa"]})
df1
```

```python
df2 = pd.DataFrame({"id": [103, 104, 105],
                    "first_name": ["ccile", "ludivine", "kevin"],
                    "last_name": ["hevert", "masseau", "leroi"]})
df2
```

To concatenate two DataFrames, you need to use the `concat` function of pandas, **beware** of the axis parameter, which determines how to concatenate (in rows or columns)

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
df_concat = pd.concat([df2,df1], axis=0)
df_concat
```

Reordering the names of the indices

```python
df_concat = df_concat.reset_index(drop=True)
df_concat
```

```python
pd.concat([df1, df2], axis=1) # With opposite axis
```

### join

```python
df_concat
```

```python
df3 = pd.DataFrame({"id": [103, 100, 101, 112],
                    "age": [28, 42, 87, 55]})
df3
```

**Inner join**, keep only the ids common to both DataFrame

```python
df_concat
```

```python
pd.merge(df_concat, df3, on="id", how="left") # you can have multiple columns
```

**Left join**, keep only the ids of the left DataFrame and we mark the correspondence if there is one!

```python
pd.merge(df_concat, df3, on="id", how="left")
```

**Outer join**, keep the union of the ids of the left and right DataFrame with correspondence when it is possible

```python
pd.merge(df_concat, df3, on=["id"], how="outer")
```
